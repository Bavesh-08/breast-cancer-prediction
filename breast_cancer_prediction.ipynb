```python
# importing the dependencies

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
# collecting the data

cancer_data=pd.read_csv('/content/breast_cancer.csv')
cancer_data.head()

```

```python
# getting the shape of the data

cancer_data.shape
```

```python
# getting info from the data

cancer_data.info()
cancer_data.isnull().sum()
```

```python
# getting the average of all the features

cancer_data.describe()
```

```python
# sepration of data in diagnosis

plt.figure(figsize=(6,4))
sns.countplot(x='diagnosis',data=cancer_data)
plt.title("stages of cancer")
plt.xlabel("diagnosis")
plt.ylabel("count")
plt.show()
```

```python
# spliting the data into features and label

x=cancer_data.drop(columns='diagnosis',axis=1)
y=cancer_data['diagnosis']
```

```python
print(x)
print(y)
```

```python
# spliting the data for training and testing

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2)
```

```python
# training the models

models=[LogisticRegression(),SVC(),RandomForestClassifier()]

for model in models:
  model.fit(x_train,y_train)
```

```python
# let mave the model predict on known data and finding the accuracy

for model in models:
  train_pred=model.predict(x_train)
  accuracy_train=accuracy_score(y_train,train_pred)
  print(f"{model}:{accuracy_train}")
```

```python
# let make the model predict on new data

for model in models:
  test_pred=model.predict(x_test)
  accuracy_train=accuracy_score(y_test,test_pred)
  print(f"{model}:{accuracy_train}")
```

```python
# confussion matrix

cls=confusion_matrix(y_test,test_pred)
sns.heatmap(cls,annot=True)
plt.show()
```
